```python
#分析收入
raw_data3 = pd.read_csv('sample_taxi.csv')
```

```python
raw_data3=raw_data3.sort_values(by=['taxi_id','time'])
raw_data3
```

```python
raw_data3=raw_data3[-((raw_data3['is_passenger'].shift(-1) == raw_data3['is_passenger'].shift())&
(raw_data3['is_passenger'].shift(-1) != raw_data3['is_passenger'])&
(raw_data3['taxi_id'].shift(-1) == raw_data3['taxi_id'].shift())&
(raw_data3['taxi_id'].shift(-1) == raw_data3['taxi_id']))]
```

```python
raw_data3.loc[:,'time'] = pd.to_datetime(raw_data3.loc[:,'time'])
raw_data3
```

```python
raw_data3['is_passenger2'] = raw_data3['is_passenger'].shift()
raw_data3['taxi_id2'] = raw_data3['taxi_id'].shift()

raw_data3['lon2'] = raw_data3['lon'].shift()
raw_data3['lat2'] = raw_data3['lat'].shift()
raw_data3['time2'] = raw_data3['time'].shift()

raw_data3['StatusChange'] = raw_data3['is_passenger2']-raw_data3['is_passenger']


raw_data3 = raw_data3[raw_data3['is_passenger2']==1]
raw_data3['orderid'] = raw_data3['StatusChange'].cumsum()
```

```python
#
table2 = raw_data3[['orderid','lon','lat','time','speed','StatusChange']]
table2.head(100)
```

```python
#计算里程

#定义计算路径长度函数

from math import pi
import numpy as np

def getdistance(lon1, lat1, lon2, lat2): # 经度1，纬度1，经度2，纬度2 （十进制度数）输入为DataFrame的列
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # 将十进制度数转化为弧度
    lon1, lat1, lon2, lat2 = map(lambda r:r*pi/180, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(a**0.5) 
    r = 6371 # 地球平均半径，单位为公里
    return c * r * 1000
```

```python
table3 = table2.copy()
table3['lon1'] = table3['lon'].shift(-1)
table3['lat1'] = table3['lat'].shift(-1)
table3['orderid1'] = table3['orderid'].shift(-1)
table3['time1'] = table3['time'].shift(-1)
table3 = table3[(table3['orderid1'] == table3['orderid'])]

lon1 = table3['lon']
lat1 = table3['lat']
lon2 = table3['lon1']
lat2 = table3['lat1']
table3['distance'] = getdistance(lon1, lat1, lon2, lat2)
```

```python
table3
```

```python
table3['interval'] = (table3['time1']-table3['time']).apply(lambda r:r.seconds)
table3
```

```python
#有一定偏差，所以 waittime 就是distance 小于20m的都算是等待时间
orderlenth = table3[['orderid','distance']].groupby('orderid').sum().reset_index()

waittime = table3[table3['distance']<=20][['orderid','interval']].groupby('orderid').sum()

table3 = pd.merge(orderlenth,waittime,on = 'orderid',how='left')
table3.head(10)
```

```python

#将每一个出行的起点与终点提取出来
o = raw_data3.iloc[:1].append(raw_data3[raw_data3['StatusChange']==1])
d = raw_data3[(raw_data3['StatusChange']==1).shift(-1).fillna(False)]
table4 = o.append(d).sort_values(by = ['orderid','time'])[['orderid','taxi_id','time','lat','lon']]
table4.head(5)
```

```python
#加一列isd，如果该行为起点，则isd=0，如果该行为终点，则isd=1
table4['isd'] = [i%2 for i in range(len(table4))]

```

```python
table4.head(5)
```

```python

# 把O与D的信息放在同一行
table4['Etime'] = table4['time'].shift(-1)
table4['ELng'] = table4['lon'].shift(-1)
table4['ELat'] = table4['lat'].shift(-1)
table4 = table4[(table4['isd']==0)&(-table4['Etime'].isnull())]
```

```python
table4
table9=pd.DataFrame(table4['taxi_id','time','lat','lon','Etime','ELng','ELat'])
table9.to_csv(r'Taxi_self_od.csv',index = None)
```

```python
#连接表3与表4成为表5
table5 = pd.merge(table4,table3,on = 'orderid',how = 'left')
table5.head(5)
```

```python
#每天的23时至次日凌晨6时为夜间
table5['isnight'] = (table5['time'].apply(lambda r:r.hour)<6)|(table5['time'].apply(lambda r:r.hour)>=23)
table5.tail(5)
```

```python

#起步价
table5['起步价'] = table5['isnight']*(16-11)+11
#里程价
table5['里程价'] = ((table5['distance']-2000)>0)*(table5['distance']-2000)*2.4/1000
#返空费
table5['返空费'] = table5['isnight']*((table5['distance']-25000)>0)*((table5['distance']-25000)*2.4*0.3/1000)

#夜间附加费(里程价)
table5['夜间附加费'] = table5['isnight']*((table5['distance']-2000)>0)*(table5['distance']-2000)*2.4*0.2/1000

#候时费
table5['候时费'] = table5['interval']/60*0.8

#最终价格
table5['price'] = table5['起步价'] + table5['里程价'] + table5['返空费'] + table5['夜间附加费'] + table5['候时费']
table5.head(5)
```

```python
#筛选掉没有计算出来价格的数据
table5 = table5[-table5['price'].isnull()]
table5
```
